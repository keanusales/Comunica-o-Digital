Desenvolva um programa (em Phyton) que processe as seguintes sequências de bits:

1000000001010011

1110100101000010

O programa deverá codificar essas sequências, conforme as diretrizes a seguir:

Cada equipe deve implementar três diferentes esquemas de codificação, de acordo com a distribuição abaixo;

Códigos: NRZ, B8ZS (Bipolar with Eight Zeros Substitution), Manchester diferencial;

O programa deve exibir graficamente, tanto as sequências originais quanto os resultados obtidos após cada codificação.

```python
from matplotlib.pyplot import figure
from itertools import product
from math import log2, floor
```

```python
def gen_img(funcname: str,
            sequence: int,
            dom: list[int],
            img: list[int]):
  fig = figure(None, (12, 2), 150)
  axe = fig.subplots()
  axe.set_title(f"{funcname} - {bin(sequence)}")
  axe.set_xlabel("Tempo")
  axe.set_ylabel("Voltagem")
  axe.step(dom, img, where = "post")
  axe.grid(True)
  fig.tight_layout(pad = 1)
```

```python
def nrz(data: int):
  order = floor(log2(data)) + 1
  dom = list(range(order))
  img = [5 if data & (1 << i) else -5
    for i in reversed(range(order))]
  dom.append(order)
  img.append(img[-1])
  return dom, img
```

```python
def b8zs(data: int):
  order = floor(log2(data)) + 1
  dom = list(range(order))
  img = []
  ones, even, zeros = 0, 0, 0
  for i in reversed(range(order)):
    if data & (1 << i):
      img.extend(([0] * zeros) + [5 if ones else -5])
      ones ^= 1
      zeros = 0
    else:
      zeros += 1
      if zeros == 8:
        if even:
          img.extend([0, 0, 0, 5, -5, 0, -5, 5])
        else:
          img.extend([0, 0, 0, -5, 5, 0, 5, -5])
        even ^= 1
        zeros = 0
  img.extend([0] * zeros)
  dom.append(order)
  img.append(img[-1])
  return dom, img
```

```python
def dif_manchester(data: int):
  order = floor(log2(data)) + 1
  dom = [e / 2 for e in range(order * 2)]
  prev, img = 0, []
  for i in reversed(range(order)):
    prev ^= bool(data & (1 << i))
    img.extend(map(lambda x: 5 if x
      else -5, (prev ^ 1, prev)))
  dom.append(order)
  img.append(img[-1])
  return dom, img
```

```python
funcs = (nrz, b8zs, dif_manchester)
seqs = (0b1000000001010011, 0b1110100101000010)
for func, seq in product(funcs, seqs):
  dom, img = func(seq)
  gen_img(func.__name__, seq, dom, img)
```
